```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set_style('darkgrid')


def illustrate_learning(a=1.3, b=0.5, n=30, r=0.5, make_prediction=False):
    np.random.seed(42)
    def f(x): return a * x + b + r * np.random.rand(len(x))
    
    x = np.random.rand(n)
    fig, ax = plt.subplots()
    ax.set_xlim([-0.05, 1.15])
    sns.regplot(x, f(x))
    if make_prediction:
        x_ = np.array(max(x) + 0.1).reshape(1,1) # funky
        plt.plot(x_, f(x_), 'ro')
    plt.xlabel('x', fontsize=15)
    plt.ylabel('f(x)', fontsize=15)
    plt.show()
```

## Machine learning

> "The field of study that allows computers to learn without being explicitly programmed." 

\- Arthur Samuel  -  (*He coined the term "machine learning" in 1959*)

### What does this _learning_ mean?

> "A computer program is said to learn from experience E with respect some class of tasks T and performance measure P , if its performance at tasks in T , as measured by P , improves with experience E." 

\- Tom Mitchell

Example: the self-driving car

- **E** = the experience of driving a car around town
- **T** = the task of driving a car to some destination
- **P** = reaching the destination on time and without accidents  

### In practice

Algorithms that learn from data to make predictions or do inference.

Examples:

- Predicting credit risk of banking customers
- Classifying emails as spam or legitimate 
- Identifying human faces in photographs
- Predicting maintenance needs for elevators

### The learning problem

Goal: learn the relationships that capture the systematic information present in our data.

Model should learn so that it can generalize from the data.

```python
illustrate_learning()
```

### The learned model

The learned model is used for:
- predicting responses of yet-unseen data points 
- understanding the relationships in the data

```python
illustrate_learning(make_prediction=True)
```

### The learned model

> "All models are wrong, but some are useful." 

\- George E. P. Box

- Our learned models will always be 'wrong', as they are a simplification of reality
- However, derived relationships and predictions can be indicative of the 'truth'

If our learned models are wrong, how can they be useful?

## What is scikit-learn?
